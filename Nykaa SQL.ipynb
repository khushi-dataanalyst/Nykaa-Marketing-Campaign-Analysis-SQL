Nykaa Marketing Campaign Performance Analysis Using SQL

Project Overview

This project analyzes Nykaa marketing campaign performance using SQL.
The objective is to evaluate campaign effectiveness, customer engagement,
conversion rates, ROI, and revenue generation across different marketing channels.

Import Required Libraries

```python
import pandas as pd
import sqlite3
```

Load Dataset

```python
df = pd.read_csv("nykaa_campaign_data.csv")

df.head()
```

Basic Data Check

```python
df.info()

df.isnull().sum()

df.describe()
```

Create SQLite Database

```python
conn = sqlite3.connect("NykaaMarketing.db")

cursor = conn.cursor()
```

Business Questions

1. What is the total revenue generated?
2. Which marketing channels drive the highest revenue?
3. Which campaign types have the highest ROI?
4. Which customer segments contribute most revenue?
5. Which target audiences convert best?
6. What are the top-performing campaigns?

Import Data into Table

```python
df.to_sql(
    "marketing_campaign",
    conn,
    if_exists="replace",
    index=False
)

print("Data Imported Successfully")
```

Verify Data

```python
query = """
SELECT *
FROM marketing_campaign
LIMIT 5
"""

pd.read_sql(query, conn)
```

Total Revenue

```python
query = """
SELECT
SUM(revenue) AS Total_Revenue
FROM marketing_campaign
"""

pd.read_sql(query, conn)
```

Total Campaign Spend

```python
query = """
SELECT
SUM(acquisition_cost) AS Total_Spend
FROM marketing_campaign
"""

pd.read_sql(query, conn)
```

Overall ROAS

```python
query = """
SELECT
ROUND(SUM(revenue)/SUM(acquisition_cost),2) AS Overall_ROAS
FROM marketing_campaign
"""

pd.read_sql(query, conn)
```

Revenue by Channel

```python
query = """
SELECT
channel_used,
SUM(revenue) AS Revenue
FROM marketing_campaign
GROUP BY channel_used
ORDER BY Revenue DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

Best Campaign Type

```python
query = """
SELECT
campaign_type,
SUM(revenue) AS Revenue
FROM marketing_campaign
GROUP BY campaign_type
ORDER BY Revenue DESC
"""

pd.read_sql(query, conn)
```

Conversion Rate by Campaign Type

```python
query = """
SELECT
campaign_type,
ROUND(
    SUM(conversions) * 100.0 / SUM(clicks),
    2
) AS Conversion_Rate
FROM marketing_campaign
GROUP BY campaign_type
ORDER BY Conversion_Rate DESC
"""

pd.read_sql(query, conn)
```

Top 10 Revenue Generating Campaigns

```python
query = """
SELECT
campaign_id,
campaign_type,
channel_used,
revenue
FROM marketing_campaign
ORDER BY revenue DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

Customer Segment Analysis

```python
query = """
SELECT
customer_segment,
SUM(revenue) AS Revenue,
SUM(conversions) AS Conversions,
ROUND(
SUM(revenue)*1.0/SUM(conversions),
2
) AS Revenue_Per_Conversion
FROM marketing_campaign
GROUP BY customer_segment
ORDER BY Revenue DESC
"""

pd.read_sql(query, conn)
```

Target Audience Analysis

```python
query = """
SELECT
target_audience,
SUM(revenue) AS Revenue
FROM marketing_campaign
GROUP BY target_audience
ORDER BY Revenue DESC
"""

pd.read_sql(query, conn)
```

Campaign Type ROI

```python
query = """
SELECT
campaign_type,
ROUND(
SUM(revenue)/SUM(acquisition_cost),
2
) AS ROI
FROM marketing_campaign
GROUP BY campaign_type
ORDER BY ROI DESC
"""

pd.read_sql(query, conn)
```

Highest Conversion Campaign Types

```python
query = """
SELECT
campaign_type,
SUM(conversions) AS Total_Conversions
FROM marketing_campaign
GROUP BY campaign_type
ORDER BY Total_Conversions DESC
"""

pd.read_sql(query, conn)
```
